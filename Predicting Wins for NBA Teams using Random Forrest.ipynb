```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df_team=pd.read_csv('team_data.csv')
df_team.describe()
```

```python
df_team = df_team[df_team.columns.drop(list(df_team.filter(regex='rid')))]
```

```python
df_team.describe()
```

```python
dataTypeSeries=df_team.dtypes
dataTypeSeries
```

```python
df_team['home'] = np.where(df_team['team']== df_team['home_team'], '1', '0')
df_team['home']
```

```python
df_team = df_team.drop(['home_team', 'opp_team', 'team'], axis=1)
```

```python
df_team = df_team.drop(['score', 'year', 'coach_black', 'coachid', 'white', 'fracwhite', 'ref1_id', 'ref2_id', 'ref3_id'], axis=1)
```

```python
df_team = df_team.drop(['edate', 'sample', 'min', 'blackmin', 'whitemin', 'berri', 'out_cont', 'gameid', 'teamgame', 'tm_yr'], axis=1)
```

```python
df_team = df_team.drop(['televised', 'ot', 'opp_score', 'diff_score', 'margin', 'opp_sample', 'diff_sample',], axis=1)
```

```python
df_team = df_team.drop(['opp_home', 'diff_home', 'opp_ot', 'diff_ot', 'opp_min', 'diff_min',], axis=1)
```

```python
df_team = df_team[df_team.columns.drop(list(df_team.filter(regex='white')))]
```

```python
df_team = df_team[df_team.columns.drop(list(df_team.filter(regex='black')))]
```

```python
df_team = df_team[df_team.columns.drop(list(df_team.filter(regex='starter')))]
```

```python
df_team = df_team[df_team.columns.drop(list(df_team.filter(regex='berri')))]
```

```python
df_team = df_team[df_team.columns.drop(list(df_team.filter(regex='points')))]
```

```python
df_team = df_team.drop(['opp_win', 'diff_win', 'opp_margin', 'diff_margin', 'min_sum_min',], axis=1)
```

```python
# Imputation
df_team1 = df_team
imputer = SimpleImputer()
df_team1 = pd.DataFrame(imputer.fit_transform(df_team1))
df_team1.columns = df_team.columns
df_team1.head()
```

```python
y=df_team1['win']
X=df_team1.drop(['win'], axis=1)
```

```python
X.head()
if 'win' in X.columns:
    print("Yes")
else:
    print("No")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_dict = {"n_estimators":[100,500,1000,2500],
              "max_depth":[1,3,5,10]}
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
gs = GridSearchCV(rf, param_dict, scoring = "f1_macro", 
                  n_jobs = -1, verbose = 2)
gs.fit(X_train, y_train)
print("Best Params: {}".format(gs.best_params_))
```

```python
rf1= RandomForestClassifier(max_depth=10, n_estimators=1000)
rf1 = rf1.fit(X_train, y_train)
forest_pred = rf1.predict(X_test)
```

```python
from sklearn.metrics import f1_score, classification_report
print(classification_report(y_test, forest_pred))
from sklearn.metrics import confusion_matrix
y_pred = gs.predict(X_test)
conmat = confusion_matrix(y_test, y_pred)
print(conmat)
```
